```python
import pandas as pd
X = numpy.concatenate((full_x, bank_x), axis=0)
y = numpy.concatenate((full_y, bank_y), axis=0)
show_y = pd.DataFrame(set(y)).T
show_y
```

## t-SNE or UMAP

```python
import matplotlib.pyplot as plt
from scipy.spatial import ConvexHull
filtr_bank, filtr_full = numpy.isin(y, bank_y), numpy.isin(y, full_y)
filtrs_bank = [y == y_l for y_l in bank_label]
filtrs_full = [y == y_l for y_l in full_label]
```

```python
from sklearn.manifold import TSNE
from umap import UMAP
# D_reduce = TSNE(n_components=2, random_state=1)
D_reduce = UMAP(n_components=2, random_state=22, min_dist=0.7)

X_dd = D_reduce.fit_transform(X)
```

### ID 1

```python
import matplotlib.pylab as plt

fig, ax = plt.subplots(1, 3, figsize=(30, 10))

filtr_b = filtrs_bank[0]
filtr_f = filtrs_full[0]

# 聚类
from sklearn.cluster import KMeans, OPTICS, HDBSCAN, DBSCAN
from sklearn.cluster import OPTICS
from sklearn.cluster import SpectralClustering

cluster0 = DBSCAN()
cluster1 = DBSCAN(eps=1.5)
f_l = cluster0.fit_predict(X_dd[filtr_f])
b_l = cluster1.fit_predict(X_dd[filtr_b])

clus = pd.concat([pd.DataFrame(set(f_l)), pd.DataFrame(set(b_l))], axis=0)

ax[0].scatter(X_dd[filtr_f, 0], X_dd[filtr_f, 1], color=(48/255, 129/255, 146/255), s=8, label="原始数据")
ax[0].scatter(X_dd[filtr_b, 0], X_dd[filtr_b, 1], color=(199/255, 77/255, 38/255), s=8, label="筛选所得数据")

f_clus = []
b_clus = []

for clu_idx in set(f_l):
    f_clu = X_dd[filtr_f][f_l==clu_idx]
    ax[1].scatter(f_clu[:, 0], f_clu[:, 1])
    if clu_idx != -1:
        f_hull = ConvexHull(f_clu) 
        ax[1].plot(f_clu[f_hull.vertices, 0], f_clu[f_hull.vertices, 1], 'r--')
        ax[1].plot([f_clu[f_hull.vertices, 0][0], f_clu[f_hull.vertices, 0][-1]],
                    [f_clu[f_hull.vertices, 1][0], f_clu[f_hull.vertices, 1][-1]],
                    'r--')
        f_clus.append(f_clu[f_hull.vertices])

    
for clu_idx in set(b_l):
    b_clu = X_dd[filtr_b][b_l==clu_idx]
    ax[2].scatter(b_clu[:, 0], b_clu[:, 1])
    if clu_idx != -1:
        b_hull = ConvexHull(b_clu)
        ax[2].plot(b_clu[b_hull.vertices, 0], b_clu[b_hull.vertices, 1], 'r--')
        ax[2].plot([b_clu[b_hull.vertices, 0][0], b_clu[b_hull.vertices, 0][-1]],
                    [b_clu[b_hull.vertices, 1][0], b_clu[b_hull.vertices, 1][-1]],
                    'r--')
        b_clus.append(b_clu[b_hull.vertices])

title = ['(A)数据筛选结果', '(B)原始数据凸包', '(C)筛选数据凸包']
for i in range(3):
    ax[i].set_xlabel('umap-1', fontsize=20)
    ax[i].set_ylabel('umap-2', fontsize=20)
    ax[i].set_xlim(-18, 25)
    ax[i].set_ylim(-10, 25)
    ax[i].tick_params(labelsize=20)
    ax[i].set_title(title[i], fontsize=25)
    ax[0].legend(fontsize=15)

plt.show()
```
