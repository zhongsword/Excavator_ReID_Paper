### 导入数据

```python
from dataLoader import Datasets
import numpy

datasets = Datasets("../feature_bank.json")
bank_x = datasets.data
bank_y = datasets.target
bank_y = [x+'bank' for x in bank_y.tolist()]
bank_y = numpy.array(bank_y)
print(len(bank_y))
```

```python
from dataLoader import FullDatasets

datasets = FullDatasets("../feature_map.json")
full_x = datasets.data
full_y = datasets.target
```

```python
import pandas as pd
X = numpy.concatenate((full_x, bank_x), axis=0)
y = numpy.concatenate((full_y, bank_y), axis=0)
show_y = pd.DataFrame(set(y))
show_y
```

### t-SNE

```python
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2, random_state=1)
X_tsne = tsne.fit_transform(X)
```

```python
import matplotlib.pyplot as plt
from matplotlib.cm import get_cmap

filtr_bank, filtr_full= numpy.isin(y, bank_y), numpy.isin(y, full_y) 
filtr_1, filtr_2, filtr_3, filtr_8, filtr_15, filtr_bank_17 = \
    y == '1.0', y == '2.0',y == '3.0',y == '8.0',y == '15.0',y == '17.0'
filtr_1_b, filtr_2_b, filtr_3_b, filtr_8_b, filtr_15_b, filtr_bank_17_b = \
    y == '1.0bank', y == '2.0bank',y == '3.0bank',y == '8.0bank',y == '15.0bank',y == '17.0bank'

colormap = get_cmap('plasma')

# ax[0].scatter(X_tsne[filtr_full, 0], X_tsne[filtr_full, 1], color='blue')
# ax[0].scatter(X_tsne[filtr_bank, 0], X_tsne[filtr_bank, 1], color='red')
```

### ID 1

```python
fig, ax = plt.subplots(2, 1, figsize=(11, 20))

ax[0].scatter(X_tsne[filtr_1, 0], X_tsne[filtr_1, 1], color='blue')
ax[0].scatter(X_tsne[filtr_1_b, 0], X_tsne[filtr_1_b, 1], color='red')


for x, color in zip(X_tsne[filtr_1_b], colormap(numpy.linspace(0, 1,len(X_tsne[filtr_1_b])))):
    ax[1].scatter(x[0], x[1], color=color)

sc = ax[1].scatter(X_tsne[filtr_1_b, 0], X_tsne[filtr_1_b, 1], 
                   c=colormap(numpy.linspace(0, 1, len(X_tsne[filtr_1_b]))),
                   cmap=colormap)
cbar = fig.colorbar(sc, ax=ax[1], orientation='vertical', fraction=0.02, pad=0.001)
cbar.set_label('Time')
 
ax[0].set_xlabel('t-SNE1')
ax[0].set_ylabel('t-SNE2')
for i in range(2):
    ax[i].set_xlim(-120, 120)
    ax[i].set_ylim(-120, 120)
    ax[i].legend()

plt.title("t-SNE of feature bank")
plt.show()
```
