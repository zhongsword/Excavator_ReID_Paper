```python
# YOLO seg
yds_train = pd.io.parsers.read_csv("/home/zlj/ultralytics/runs/segment/train3/results.csv", header=None)
yds_train_box = yds_train[1].values[1:].astype(np.float32)
yds_train_seg = yds_train[2].values[1:].astype(np.float32)
yds_train_cls = yds_train[3].values[1:].astype(np.float32)
yds_val_box = yds_train[13].values[1:].astype(np.float32)
yds_val_seg = yds_train[14].values[1:].astype(np.float32)
yds_val_cls = yds_train[15].values[1:].astype(np.float32)
```

```python
# YOLO
yd_train = pd.io.parsers.read_csv("/home/zlj/ultralytics/runs/detect/train29/results.csv", header=None)
yd_train_box = yd_train[1].values[1:].astype(float)
yd_train_cls = yd_train[2].values[1:].astype(float)
yd_val_box = yd_train[8].values[1:].astype(float)
yd_val_cls= yd_train[9].values[1:].astype(float)
yd_train_box
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(2,6, figsize=(20, 6))

#
ax[0][0].plot(range(len(yd_train_box)), yd_train_box)
ax[1][0].plot(range(len(yd_val_box)), yd_val_box)
ax[0][0].set_title("(A)YOLOv8检测框训练损失", fontsize=18)
ax[1][0].set_title("(B)YOLOv8检测框验证损失", fontsize=18)

#
ax[0][1].plot(range(len(yd_train_cls)), yd_train_cls)
ax[1][1].plot(range(len(yd_val_cls)), yd_val_cls)
ax[0][1].set_title("(C)YOLOv8分类训练损失", fontsize=18)
ax[1][1].set_title("(D)YOLOv8分类验证损失", fontsize=18)

#
ax[0][2].plot(range(len(yds_train_seg)), yds_train_seg)
ax[1][2].plot(range(len(yds_val_seg)), yds_val_seg)
ax[0][2].set_title("(E)YOLOv8-seg掩码训练损失", fontsize=18)
ax[1][2].set_title("(F)YOLOv8-seg掩码验证损失", fontsize=18)

#
ax[0][3].plot(range(len(yds_train_cls)), yds_train_cls)
ax[1][3].plot(range(len(yds_val_cls)), yds_val_cls)
ax[0][3].set_title("(G)YOLOv8-seg分类训练损失", fontsize=18)
ax[1][3].set_title("(H)YOLOv8-seg分类验证损失", fontsize=18)

#
ax[0][4].plot(range(len(rsnt_train)), rsnt_train)
ax[1][4].plot(range(len(rsnt_val)), rsnt_val)
ax[0][4].set_title("(I)Resnet50训练损失", fontsize=18)
ax[1][4].set_title("(J)Resnet50验证损失", fontsize=18)

#
ax[0][5].plot(range(len(cl_train)), cl_train)
ax[1][5].plot(range(len(cl_val)), cl_val)
ax[0][5].set_title("(K)CNN_LSTM训练损失", fontsize=18)
ax[1][5].set_title("(L)CNN_LSTM验证损失", fontsize=18)

# # train_loss
# ax[0].plot(idx, train[:20, 0])

# # train_acc
# ax[1].plot(idx, train[:20, 1])

# # val_loss
# ax[2].plot(idx, val[:20, 0])

# # val_acc
# ax[3].plot(idx, val[:20, 1])

# label = ['train LOSS', 'train ACCURACY', 'val LOSS', 'val ACCURACY']

# for i in range(4):
#     ax[i].set_xlabel("Epoch",fontsize=15)
#     ax[i].tick_params(axis='both', which='major', labelsize=15)
ticks = ax[0][5].get_xticks()
x_set = [f'{int(tick*2)}' for tick in ticks]
#     ax[i].set_xticklabels(x_set)
#     ax[i].set_title(label[i], fontsize=20)

ax[0][5].set_xticklabels(x_set)
ax[1][5].set_xticklabels(x_set)

for i in ax:
    for j in i:
        j.tick_params(axis='both', which='major', labelsize=12)

fig.tight_layout()
plt.show()
```
